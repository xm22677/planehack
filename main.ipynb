```python
import torch.nn as nn
```

```python

class LSTM_Model(nn.Module):
    def __init__(self,inputDim=10,hiddenDim=32,layer=1,inputDim2=32,hiddenDim2=16,layer2=1,):
        super(LSTM_Model).__init__()
        self.inputDim = inputDim
        self.hiddenDim = hiddenDim
        self.layers = layer
        self.inputDim2 = inputDim2
        self.hiddenDim2 = hiddenDim2
        self.layers2 = layer2
        
        self.LSTM1 = nn.LSTM(
                input_size=self.inputDim,
                hidden_size=self.hiddenDim,
                num_layers=self.layers,
                bias=True,
                batch_first=True,
                bidirectional=True,

                )
        self.LSTM2 = nn.LSTM(
                input_size=self.inputDim*2,
                hidden_size=self.hiddenDim2,
                num_layers=self.layers2,
                bias=True,
                batch_first=True,
                bidirectional=True,

                )
        self.L1 = nn.Linear(32,16,bias=False)
        self.A1 = nn.ReLU()
        self.L2 = nn.Linear(16,8,bias=False)
        self.A2 = nn.ReLU()
        self.L3 = nn.Linear(8,1,bias=False)

    def forward(self,X):
        X = self.LSTM1(X)
        X = self.LSTM2(X)
        X = self.A1(self.L1(X))
        X = self.A2(self.L2(X))
        return self.L3(X)


        
        

            


```
